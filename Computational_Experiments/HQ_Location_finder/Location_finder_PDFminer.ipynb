```python
import os
import pandas as pd
from transformers import pipeline
from pdfminer.high_level import extract_text

def extract_text_from_pdf(file_path):
    return extract_text(file_path)

def extract_company_location(text):
    question = "What is the location of the company?"
    model_name = "distilbert-base-cased-distilled-squad"
    model_revision = "626af31"
    nlp = pipeline("question-answering", model=model_name, revision=model_revision)
    answer = nlp(question=question, context=text)
    return answer['answer']

pdf_folder = '/Users/amin/Desktop/33/test'

pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

pdf_texts = {pdf_file[:-4]: extract_text_from_pdf(os.path.join(pdf_folder, pdf_file))
             for pdf_file in pdf_files}

data = []

for filename, text in pdf_texts.items():
    location = extract_company_location(text)
    data.append({'filename': filename, 'location': location})

df = pd.DataFrame(data)
print(df)
```
