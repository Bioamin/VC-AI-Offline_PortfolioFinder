```python
import pandas as pd
import os

os.chdir("/Users/amin/Desktop/Symbiosis_Amin/Extracted_Data_Files")

# Read the CSV file
df = pd.read_csv('Data.csv')

# List of columns you want to calculate the frequency for
columns = ['Issue', 'Method', 'location_area', 'Diag_Treat']

# Calculate the frequency and percentage for each column
for col in columns:
    print(f"Frequency and percentage for {col}:")
    
    # Calculate frequency
    frequency = df[col].value_counts()
    print(frequency)

    # Calculate percentage
    percentage = (frequency / df.shape[0]) * 100
    print("\nPercentage:")
    print(percentage)
    print("\n")
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('Data.csv')

# Define the intervals and labels
bins = [0, 3, 10, 20, float('inf')]
labels = ['[0, 3]', '[3, 10]', '[10, 20]', '[20+]']

# Categorize the data into intervals
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels)

# Get the frequency and percentage for each interval
age_group_counts = data['Age_group'].value_counts()
age_group_percentage = data['Age_group'].value_counts(normalize=True) * 100

# Print the frequency and percentage
print("Frequency of each age group:")
print(age_group_counts)
print("\nPercentage of each age group:")
print(age_group_percentage)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV filea
data = pd.read_csv('Data.csv')

# Get the frequency of each category in the columns
computationality_counts = data['Computationality'].value_counts()
clinical_counts = data['Clinical'].value_counts()
age_counts = data['Age'].value_counts()

# Print the frequency counts
print("Frequency of each category in Computationality column:")
print(computationality_counts)
print("\nFrequency of each category in Clinical column:")
print(clinical_counts)
print("\nFrequency of each category in Age column:")
print(age_counts)

# Plot the bar charts
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

ax[0].bar(computationality_counts.index, computationality_counts.values)
ax[0].set_title('Computationality Frequencies')
ax[0].set_xlabel('Computationality')
ax[0].set_ylabel('Frequency')

ax[1].bar(clinical_counts.index, clinical_counts.values)
ax[1].set_title('Clinical Frequencies')
ax[1].set_xlabel('Clinical')
ax[1].set_ylabel('Frequency')

ax[2].bar(age_counts.index, age_counts.values)
ax[2].set_title('Age Frequencies')
ax[2].set_xlabel('Age')
ax[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show
```

```python
import pandas as pd

# Read the CSV file
data = pd.read_csv('Data.csv')

# Define the intervals and labels
bins = [0, 2, 4, 6, 8, 10]
labels = ['[0, 2]', '[2, 4]', '[4, 6]', '[6, 8]', '[8, 10]']

# Categorize the data into intervals
data['Computationality_group'] = pd.cut(data['Computationality'], bins=bins, labels=labels, include_lowest=True)

# Get the frequency and percentage for each interval
computationality_group_counts = data['Computationality_group'].value_counts().sort_index()
computationality_group_percentage = data['Computationality_group'].value_counts(normalize=True).sort_index() * 100

# Print the frequency and percentage
print("Frequency of each computationality group:")
print(computationality_group_counts)
print("\nPercentage of each computationality group:")
print(computationality_group_percentage)
```

```python
import pandas as pd

# Read the Data.csv file
data = pd.read_csv('Data.csv')

# Define the intervals and labels
bins = [0.0, 0.1, 1, 2, 3]
labels = ['[0.0, 0.1]', '[0.1, 1]', '[1, 2]', '[2, 3]']

# Calculate the frequency and frequency percentage
data['interval'] = pd.cut(data['Clinical'], bins=bins, labels=labels, include_lowest=True)
frequency = data['interval'].value_counts(sort=False)
frequency_percentage = data['interval'].value_counts(normalize=True, sort=False) * 100

# Create a new DataFrame with the results
result = pd.DataFrame({'Frequency': frequency, 'Frequency Percentage': frequency_percentage})
print(result)
```
