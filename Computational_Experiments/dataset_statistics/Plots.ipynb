```python
import matplotlib.pyplot as plt

data = {'Name': 50, 'Method Category': 73, 'Disease Category': 80, 'Computationality': 42, 
        'Clinical Phase': 66, 'Location': 62, 'Age': 65, 'Website': 80, 
        'Diag_Treat': 85}

# Define mate version of colors for the bars
colors = ['indianred', 'green', 'blue', 'orange', 'purple', 'saddlebrown', 'red', 'pink', 'black']

# Creating a figure and axis object
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting the bar plot with colors
ax.bar(data.keys(), data.values(), color=colors)

# Setting the y-axis limits
ax.set_ylim([0, 100])

# Enlarging the x-axis ticks
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)

# Setting the x-axis label
ax.set_xlabel('Extracted Data')

# Setting the y-axis label
ax.set_ylabel('Similarity Percentage')

# Setting the title of the plot
ax.set_title('Similarity Percentage of Extracted Data', fontsize=14)

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Cancer': 48.576214,
    'Other': 19.932998,
    'Neurological': 13.065327,
    'Vax-Infectious': 5.862647,
    'Gastro-Metabolism': 5.695142,
    'Dermatological': 2.847571,
    'Organ Health': 2.680067,
    'Mental': 1.340034
}

categories = list(data.keys())
values = list(data.values())

# Normalize values to range [0.3, 1] to make colors darker
normalized_values = 0.7 + 0.3 * (np.array(values) / max(values))

# Create blue colormap
cmap = plt.cm.get_cmap("Blues")

# Create bar plot with blue bars proportional to their values
plt.bar(categories, values, color=cmap(normalized_values))

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, ha="right", fontsize=14)
plt.yticks(fontsize=14)

# Add labels and title
plt.xlabel("Issues", size=14)
plt.ylabel("Frequency Percentage", size=14)
#plt.title("Frequency Percentage by Category")

# Display the bar plot
plt.show()

```
