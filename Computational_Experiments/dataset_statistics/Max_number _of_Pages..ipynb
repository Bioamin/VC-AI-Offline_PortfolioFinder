```python
#This code find the PDF file with the maximum number pages

import os
import PyPDF2

folder_path = "/Users/amin/Downloads/Training set"

# Get a list of all PDF files in the folder
pdf_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.pdf')]

# Loop through all PDF files and get the number of pages in each
max_pages = 0
max_pages_file = ''
for pdf_file in pdf_files:
    try:
        with open(pdf_file, 'rb') as f:
            pdf_reader = PyPDF2.PdfFileReader(f)
            if pdf_reader.isEncrypted:
                # Skip encrypted files
                continue
            num_pages = pdf_reader.getNumPages()
            if num_pages > max_pages:
                max_pages = num_pages
                max_pages_file = pdf_file
    except PyPDF2.utils.PdfReadError:
        # Skip corrupted files
        continue

# Print the PDF file with the highest number of pages
if max_pages_file:
    print(f"The PDF file with the highest number of pages is {max_pages_file}, with {max_pages} pages.")
else:
    print("No valid PDF files found.")
```
