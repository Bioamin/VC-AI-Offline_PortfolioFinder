```python
import PyPDF2
import pandas as pd
import re
import os
import glob
from transformers import pipeline

def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfFileReader(file)
        text = ''
        for page_num in range(pdf_reader.getNumPages()):
            text += pdf_reader.getPage(page_num).extractText()
    return text

def extract_year(text, nlp_model):
    question = "What year was the company founded?"
    answer = nlp_model(question=question, context=text)
    year = None
    if answer:
        year_match = re.search(r'\b(19|20)\d{2}\b', answer['answer'])
        if year_match:
            year = int(year_match.group())
    return year

folder_path = '/Users/amin/Desktop/33/test' 
pdf_files = glob.glob(os.path.join(folder_path, '*.pdf'))

data = []

nlp = pipeline('question-answering', model='distilbert-base-uncased-distilled-squad')

for pdf_file in pdf_files:
    company_name = os.path.basename(pdf_file).replace('.pdf', '')
    pdf_text = extract_text_from_pdf(pdf_file)
    founding_year = extract_year(pdf_text, nlp)
    if founding_year:
        company_age = 2023 - founding_year
        data.append([company_name, company_age])

df = pd.DataFrame(data, columns=['Company Name', 'Company Age'])
print(df)
```
