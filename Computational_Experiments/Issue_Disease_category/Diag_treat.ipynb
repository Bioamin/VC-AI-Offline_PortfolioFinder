```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

data = {'Name': 50, 'Method Category': 73, 'Disease Category': 80, 'Computationality(+-2)': 42,
        'Clinical Phase': 66, 'Location': 62, 'Age': 65, 'Website': 80, 'Human_NonHuman based': 100, 'Treat_Diag Approach': 100}

keys = list(data.keys())
values = list(data.values())

colors = cm.Greens([v/100 for v in values]) # generate colors using Greens colormap and normalized values
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(keys, values, color=colors)
ax.set_ylim(0, 100)
ax.set_ylabel('Data Extraction Accuracy Percentage', size=12)
#ax.set_title('Data Visualization')
ax.tick_params(axis='x', labelrotation=90, labelsize=15)
plt.show()
```
