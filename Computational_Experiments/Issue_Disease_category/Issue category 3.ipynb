```python
import PyPDF2
import os
from transformers import pipeline

def read_pdf(file_path):
    with open(file_path, 'rb') as f:
        pdf_reader = PyPDF2.PdfFileReader(f)
        text = ""
        for page in range(pdf_reader.getNumPages()):
            text += pdf_reader.getPage(page).extractText()
    return text

def classify_problem_area(text):
    areas = [
        "Vaccines or infectious diseases",
        "Gastrointestinal or metabolism",
        "Neurological disorders",
        "Cancer or tumors or oncology",
        "Dermatological conditions",
        "Organ health",
        "Mental health",
        "Other"
    ]

    nlp = pipeline("zero-shot-classification")
    result = nlp(text, candidate_labels=areas, multi_class=False)
    area_index = result["labels"].index(max(result["labels"], key=result["scores"].get))
    return areas[area_index]

def main():
    pdf_folder = "/Users/amin/Desktop/33/test"
    pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

    for pdf_file in pdf_files:
        file_path = os.path.join(pdf_folder, pdf_file)
        text = read_pdf(file_path)
        area = classify_problem_area(text)
        print(f"{pdf_file}: {area}")

if __name__ == "__main__":
    main()
```
