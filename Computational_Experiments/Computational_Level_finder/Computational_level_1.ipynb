```python
import os
import PyPDF2
from transformers import pipeline

def extract_text_from_pdfs(pdf_folder):
    pdf_texts = []
    
    for file in os.listdir(pdf_folder):
        if file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_folder, file)
            pdf_file = open(pdf_path, 'rb')
            pdf_reader = PyPDF2.PdfFileReader(pdf_file)
            num_pages = pdf_reader.numPages
            pdf_text = ''
            
            for page in range(num_pages):
                pdf_page = pdf_reader.getPage(page)
                pdf_text += pdf_page.extractText()
            
            pdf_texts.append(pdf_text)
            pdf_file.close()
    
    return pdf_texts

def get_computational_score(texts, model):
    question = "In the scale of 0 to 10, how computational is the company?"
    scores = []
    
    for text in texts:
        answer = model(question=question, context=text)
        try:
            score = int(answer['answer'])
        except ValueError:
            print(f"Invalid answer received: {answer['answer']}")
            score = -1  # You can assign a default value or handle this case in a different way
        
        scores.append(score)
    
    return scores

def main():
    pdf_folder = "/Users/amin/Desktop/33/test"
    pdf_texts = extract_text_from_pdfs(pdf_folder)
    
    qa_model = pipeline("question-answering")
    
    computational_scores = get_computational_score(pdf_texts, qa_model)
    
    for i, score in enumerate(computational_scores):
        print(f"Company {i + 1}: {score}")

if __name__ == "__main__":
    main()
```
